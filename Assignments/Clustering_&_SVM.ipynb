### 1. Using the dataset “tips.csv” implement k-means clustering technique on two fields “total_bill”, “tip”.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\Nirmalya Majhi\Desktop\Advanced IT Workshop\tips.csv')
print(df.head())
x = 'tip'
y = 'total_bill'
kmeans = KMeans(init="random",n_clusters=3,n_init=10,max_iter=300,random_state=42)
kmeans.fit(df[[x,y]])
print(kmeans.labels_)
plt.scatter(df[x],df[y],c=kmeans.labels_)
plt.show()
```
